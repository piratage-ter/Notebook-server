```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import pandas as pd
import geopandas as gp
import matplotlib.pyplot as plt


columns = {
    "netusers": "usagescore",
    "servers": "servsperhab",
    "Migration": "value",
    "ict_imports": "percent", 
    "ict_exports": "percent", 
    "population": "population",
    "piba" : "pibphab",
    "c_crime" : "score",
    "political_regime": "score",
    "type_political_regime" : "score_regime",
    "bsa" : "value",
    "gmi" : "score",
    "homicides" : "count",
    "literacy15" : "literacy_15",
    "literacy25_64" : "literacy24_64",
    "literacy65" : "literacy_65"
}



@interact
def display_map(name="piba", year=(1990,2020)):
    plt.close('all')
    world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))
    countries = pd.read_sql("countries", engine)
    if name == "literacy15" or name == "literacy25_64" or name == "literacy65":
        table = pd.read_sql("literacy_rate", engine)
    else:
        table = pd.read_sql(name, engine)
    #print(table.head())
    #print(countries.head())
    world["name"] = world["name"].replace({   'United States of America':'United States',
                                              'Libya':'Libyan Arab Jamahiriya',
                                              'Russia':'Russian Federation'
                                              })
    ci_table = pd.merge(countries,table, left_on = ['id'], right_on = ['country_id'])
    
    tmp = pd.merge(world,ci_table, left_on = ['name'], right_on = ['nicename'])
    tmp2 = tmp[(tmp["year"] == int(year))]
    #print(tmp2.head())
    gdf = gp.GeoDataFrame(tmp2,columns=["nicename","iso",columns[name]], crs="EPSG:4326",geometry=tmp.geometry)
    ax  = world[(world.name != "Antarctica") & (world.name!= "Fr. S. Antarctic Lands")].plot()
    ax.grid() 
    axx = gdf.plot(column = columns[name],ax=ax,cmap="Wistia", edgecolor="black",legend= True,label = 'test')
    plt.suptitle(name)
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.rcParams["figure.figsize"]=12,5
    plt.show()
    
    
    
   
```

```python
@interact
def display_map(name="piba",pays="France"):
    plt.close()
    countries = pd.read_sql("countries", engine)
    if name == "literacy15" or name == "literacy25_64" or name == "literacy65":     
        table = pd.read_sql("literacy_rate", engine)
    else:
        table = pd.read_sql(name, engine)
        
    fulltable = pd.merge(countries,table, left_on = ['id'], right_on = ['country_id'])
    cleantable = fulltable[(fulltable["nicename"]) == pays ]
    cleantable.plot(x ='year', y=columns[name], kind = 'bar')
    plt.show()
```
